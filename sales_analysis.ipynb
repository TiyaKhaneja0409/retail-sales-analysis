```python
import pandas as pd

df = pd.read_csv('train.csv', encoding='latin1')
df.head()
```

```python
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)
df.info()
```

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
region_sales
```

```python
import matplotlib.pyplot as plt

region_sales.plot(kind='bar', title='Total Sales by Region', color='steelblue')
plt.ylabel('Sales ($)')
plt.xlabel('Region')
plt.xticks(rotation=0)
plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
category_sales.plot(kind='bar', title='Total Sales by Category', color='darkorange')
plt.ylabel('Sales ($)')
plt.xticks(rotation=0)
plt.show()
```

```python
df['Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales.plot(kind='line', title='Monthly Sales Trend', figsize=(12,5), marker='o')
plt.ylabel('Sales ($)')
plt.xlabel('Month')
plt.show()
```
